```python
chromedriver = 'C:/Users\kkang/workspace/CRL/Lecture/chromedriver.exe'
headless_options = webdriver.ChromeOptions() # headless
headless_options.add_argument('headless')
headless_options.add_argument('disable-gpu') 
```

```python

for count in range(1,101,10):
    start = timeit.default_timer()
    driver=webdriver.Chrome(chromedriver,options = headless_options)
    driver.get(
            "https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=so%3Add%2Ca%3At&post_blogurl=&post_blogurl_without=&query=남이섬%20%20-맛집%20-맘들%20-분양%20-채용%20-대구여행사%20-라벤더&sm=tab_pge&srchby=title&st=date&where=post&start="+ str(count))
    last_tab = driver.window_handles[-1]
    first_tab = driver.window_handles[0]
    for i in range(1,11):
            driver.find_element_by_xpath('//*[@id="main_pack"]/div[2]/ul/li[' +str(i)+ ']/dl/dt/a').click()
            driver.switch_to.window(driver.window_handles[1])
            last_tab = driver.window_handles[-1]
            driver.switch_to.window(window_name=last_tab)
            try:
                iframe =driver.find_element_by_xpath('//*[@id="mainFrame"]')
                url = iframe.get_attribute('src')
                driver.get(str(url))                
                driver.switch_to.window(window_name=last_tab)
        
                res = requests.get(driver.current_url)
                soup = BeautifulSoup(res.content , 'html.parser')
                my_datas = soup.find_all('span')
                for data in my_datas:
                    text = data.get_text()
                    if not text in trash:
                        list_4.append(text)
                driver.close()
                driver.switch_to.window(window_name=first_tab)
    
            except (NoSuchElementException,ElementClickInterceptedException):
                driver.close() 
                driver.switch_to.window(window_name=first_tab)
                
    stop = timeit.default_timer()
    print(stop - start)
    driver.quit()
    
```
