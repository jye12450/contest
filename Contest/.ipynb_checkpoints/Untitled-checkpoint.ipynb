```python
import requests
from selenium import webdriver
import time
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,ElementClickInterceptedException
from bs4 import BeautifulSoup
import timeit
```

```python
list_2 = []
for count in range(1,1001,10):
    start = timeit.default_timer()
    driver=webdriver.Chrome("C:/Users\kkang/workspace/CRL/Lecture/chromedriver.exe")
    driver.get(
            "https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=so%3Add&post_blogurl=&post_blogurl_without=&query=ÎÇ®Ïù¥ÏÑ¨%20%2BÍ¥ÄÍ¥ë%20-ÎßòÎì§%20-Î∂ÑÏñë%20-Ï±ÑÏö©&sm=tab_pge&srchby=all&st=date&where=post&start="+ str(count))
    last_tab = driver.window_handles[-1]
    first_tab = driver.window_handles[0]
    for i in range(1,11):
            driver.find_element_by_xpath('//*[@id="main_pack"]/div[2]/ul/li[' +str(i)+ ']/dl/dt/a').click()
            driver.switch_to.window(driver.window_handles[1])
            last_tab = driver.window_handles[-1]
            driver.switch_to.window(window_name=last_tab)
            try:
                iframe =driver.find_element_by_xpath('//*[@id="mainFrame"]')
                url = iframe.get_attribute('src')
                driver.get(str(url))                
                driver.switch_to.window(window_name=last_tab)
        
                res = requests.get(driver.current_url)
                soup = BeautifulSoup(res.content , 'html.parser')
                my_datas = soup.find_all('span')
                for data in my_datas:
                    text = data.get_text()
                    if not text == "":
                        list_2.append(text)
                driver.close()
                driver.switch_to.window(window_name=first_tab)
    
            except (NoSuchElementException,ElementClickInterceptedException):
                driver.close() 
                driver.switch_to.window(window_name=first_tab)
                
    stop = timeit.default_timer()
    print(stop - start)
    
```
